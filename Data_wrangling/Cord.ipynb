### The first way to Rename unnamed columns

```python
# Rename multiple columns
df1 = df.rename(columns = {'Unnamed: 0': 'A', 'Unnamed: 1': 'B', 'Unnamed: 2': 'C', 'Unnamed: 3': 'D'})
df1.head(1)
```

```python
df1.shape
```

```python
df1.describe()
```

```python
df1.isnull().sum()
```

### Now drop the unnamed columns

```python
# Drop column A
df1.drop('A', axis=1, inplace=True)
```

```python
# Drop column B,C,D
df1.drop(['B', 'C', 'D'], axis = 1, inplace= True)
```

```python
# drop all columns from 5 to 43
df1.drop(df1.iloc[:, 5:43], axis = 1, inplace = True)
```

```python
df1.head()
```

### check for duplicate

```python
# Let's find the duplicates if any 
df[df.duplicated()]
```

```python
# Check for duplicate with 3 columns
df.duplicated(subset=['X_COORD', 'Y_COORD', 'Z_COORD']).sum
```

```python
# Drop the duplicates
df.drop_duplicates(keep='first', inplace=True)
```

```python
# Now check for duplicate 
df[df.duplicated()]
```

```python
df = df.rename(columns = {'Entity Name': 'Entity_name'})
df.head()
```

```python
# Now convert the column's header for later ease of usage if necessary
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head(1)
```

```python
# Now, save the cleaned data
df.to_csv('cleaned_Cord.csv', index=False)
```
